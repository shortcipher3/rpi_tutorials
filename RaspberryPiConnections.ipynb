## Linux
You've probably used Microsoft Windows at school, you probably know about Apple's OSX - those are two of the most popular operating systems.  You can think of an operating system as software that runs your programs that you interact with.  Linux is a third operating system - it is free software that has been written to be shared, which means you can look at the code and edit it yourself if you wanted to.

Operating Systems:
* Windows
* OSX
* Linux
* iOS
* Android

Since Linux is free it is easy for companies to adapt to their hardware/use case, it may surprise you, but you use Linux a lot.  Some devices that use Linux are:
* Raspberry Pi - uses Raspbian Linux
* Chromebook - uses Chromium (based on Linux)
* Network Router - most network routers use Linux (this is what you have at home to get to the internet)
* Websites - one estimate says that 70% of the websites you visit are hosted on Linux servers
* Android phones - Android operating system actually comes from Linux
* Vivint Smart Home devices (my work) - the cameras and sensors I work with largely are built on Linux

Not everything uses an operating system, some computers are a little more barebones, they aren't intended for general purpose use, these would be computers in electronics you have such as a camera, a modern car is estimated to have 50 computers most of which don't run an operating system, you may have heard of Arduino, which doesn't have an operating system.  These computers are often split into two groups: Real-Time Systems and Microcontrollers.

## Shell
A shell (terminal) is a way to control a computer with text commands, we have played with a few shells to get you up and running.
* crosh - the chromium shell, a limited shell for chrome users (read about the commands here: https://www.makeuseof.com/tag/terminal-commands-chromebook/ )
* bash - bourne again shell, this is probably the most popular shell available so if you want to learn more about shells it is a great one to get started on, I found a good introductory tutorial here: https://ryanstutorials.net/linuxtutorial/
* ssh - secure shell, this is a way of securely connecting with a shell to a computer over a network, typically when you connect it will start up a bash shell, or another kind of shell (sh, csh, dsh, ksh, zsh, etc)

## Exercise 1 Solutions:
Try connecting to your raspberry pi three ways
* vnc - open up vnc extension in chrome and connect to pi ip address
  1. open vnc or the chrome vnc extension
  2. enter the ip address
* ssh
  1. in chromebook open crosh `ctrl-alt-t`
  2. enter the `shell` command (chromebook must be unlocked)
  3. `ssh pi@<ip_address>`
* to a webpage - open up chrome and for the url enter the ip_address
  1. start a webserver on the raspberry pi that will serve a webpage, eg using flask
  2. in chrome for the url enter the ip address of the raspberry pi
